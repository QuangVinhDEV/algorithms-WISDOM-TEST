The HashTable class creates a hash table with size M, where each position in the table is a bucket (a list) that stores (key, value) pairs. The hash() function converts a character key into an index by taking the ASCII value of the character, subtracting 'A', and applying modulo M.

When inserting data with put(), the program first computes the hash index of the key. If the key already exists in the bucket, its value is updated; otherwise, the new (key, value) pair is appended to that bucket. The get() method searches for a key by checking the elements within the corresponding bucket and returns the associated value if found.

In the example, the string "SEARCHXAMPLE" is inserted into the hash table with the index of each character as its value. Because the table size is small (M = 5), multiple keys may map to the same index, creating collisions. These collisions are handled by storing multiple entries in the same bucket list, illustrating how separate chaining allows hash tables to store multiple keys at the same hash index while still enabling efficient search operations.

```python
class HashTable:
    def __init__(self, M=5):
        self.M = M
        self.st = [[] for _ in range(M)]

    def hash(self, key):
        return (ord(key) - ord('A')) % self.M

    def put(self, key, val):
        h = self.hash(key)
        for i, (k, v) in enumerate(self.st[h]):
            if k == key:
                self.st[h][i] = (key, val)
                return
        self.st[h].append((key, val))

    def get(self, key):
        h = self.hash(key)
        for k, v in self.st[h]:
            if k == key:
                return v
        return None


ht = HashTable()

keys = "SEARCHXAMPLE"

for i, k in enumerate(keys):
    ht.put(k, i)

print(ht.st)
print("Search E:", ht.get('E'))
```

This program builds a simple **word index** from a text file. It reads the file line by line and assigns a line number to each line. Every word is extracted, cleaned of punctuation, and stored in a dictionary (st). The dictionary acts as a **symbol table**, where the key is a word and the value is a list of line numbers where that word appears.

If a word appears multiple times, its line numbers are appended to the list. After processing the entire file, the program prints the words in sorted order along with the line numbers where they occur. If a word appears on multiple lines, it prints the range from the first line to the last line.

```python
text = """It was the best of times,
it was the worst of times,
it was the age of wisdom,
it was the age of foolishness"""

st = {}

for i, line in enumerate(text.split('\n'), 1):
    for w in line.lower().replace(',', '').split():
        st.setdefault(w, []).append(i)

for w in sorted(st):
    lines = st[w]
    print(w, f"{lines[0]}-{lines[-1]}" if len(lines) > 1 else lines[0])
```
